```python
import pandas as pd
```

I am going to learn about the New York car accident for a week from September 25th to October 1st. Using this data, we can know the area, time, and date of the most accident in a week and can be careful when passing through here.

```python
collisions = pd.read_csv("https://raw.githubusercontent.com/junphp/is362_project2/master/NYPD_Motor_Vehicle_Collisions.csv")
```

```python
collisions
```

Let's look at the most frequent time zones.

```python
top_time = collisions.groupby(['TIME']).count().sort_values('DATE',ascending=[False]).head()
top_time
```

At 0:00 we can see that the most accidents occurred.

Find out where the most frequent accidents occurred.

```python
top_area = collisions.groupby(['BOROUGH']).count().sort_values('DATE',ascending=[False]).head()
top_area
```

BROOKLYN is the place where the accident was most frequent.

Let's learn about the date of the most accident.

```python
top_date = collisions.groupby(['DATE']).count().sort_values('TIME',ascending=[False]).head()
top_date
```

On September 25, 2017, we can see that the most car accident occurred.

With the above data, we can get the following results.

```python
collisions[(collisions['DATE']=='09/25/2017')&(collisions['BOROUGH']=='BROOKLYN')&(collisions['TIME']=='0:00')].head()
```

On September 25, 2017, October 1, 2017, on Monday at 0:00, you can see that there were frequent car accidents at BROOKLYN's OCEAN AVENUE, SAINT MARKS AVENUE, HANSON PLACE and METROPOLITAN AVENUE.
