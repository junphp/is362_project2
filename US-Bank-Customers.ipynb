# US BANK CUSTOMER

```python
import pandas as pd
customer = pd.read_csv("https://raw.githubusercontent.com/junphp/is362_project2/master/US-Bank-Customers.csv")
customer
```

## 1.question

** % of female/male customers who consider themselves white or blue collar **

```python
customergroup = customer[["Job Classification","Gender"]].groupby(["Job Classification","Gender"]).agg({"Job Classification":"count"})
customergroup
```

```python
cust_pcts = customergroup.groupby(level=0).apply(lambda x: 100 * x / float(x.sum()))
cust_pcts
```

## 1.Answer

** Blue collar Female 24.30% **

** Blue collar male   75.69% **

** White collar Female 51.66% **

** White collar male   48.33% **

## 2.Question

** % of married/single folks with a good credit rating (650 and above) **

```python
customergroup2 = customer[(customer["Credit Rating"]>650) & (customer["Marriage Status"].str.contains('Single')==True)| (customer["Marriage Status"].str.contains('Married')==True)&(customer["Marriage Status"].str.contains('Married or married but separated')==False)&(customer["Marriage Status"].str.contains('Married or in consensual union')==False)&(customer["Marriage Status"].str.contains('Married, in consensual unions or separated')==False)]
customergroup2
```

```python
cust_rate=customergroup2.groupby("Marriage Status").agg({"Marriage Status":"count"})
cust_rate
```

```python
credit_pcts = customergroup2.groupby("Marriage Status").agg({"Marriage Status":"count"}).apply(lambda x: 100 * x / float(x.sum()))
credit_pcts
```

## 2.Answer

** Married 74.26% **

** Single  25.73% **

## 3.Question

** Gender with the most savings **

```python
most_saving = customer.groupby("Gender").agg({"Balance":"sum"})
most_saving
```

## 3.Answer 

** Male saving amount is great than Female saving amount **

## 4.Question

** Gender that is most employed **

```python
most_employed = customer[(customer["Employed"].str.contains('Yes')==True)].groupby("Gender").agg({"Employed":"count"})
most_employed
```

## 4.Answer

** Male is more employed than Female **
